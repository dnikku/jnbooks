```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

# Mathplotlib

```python
x = np.linspace(0, 2 * np.pi)
plt.plot(x, np.sin(x), '--', linewidth=2)
plt.plot(x, np.sin(2*x), '00', linewidth=2)
```

```python
plt.plot(x, 2*x)
```

```python
ll = np.array([1, 2, 3])
print(ll)

plt.plot(ll, 2*ll, 'r-o')
```
